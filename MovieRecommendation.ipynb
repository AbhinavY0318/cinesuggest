**NLP** **PreProcessing**

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text=text.lower()
  #removing punctuations
  text=re.sub(r'[^\a-zA-Z\s]','',text)
  #removing stopwords
  words=text.split()
  words=[word for word in words if word not in stop_words]
  #lemmatization(reducing words to their base form
  words=[lemmatizer.lemmatize(word) for word in words]
  return ' '.join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
df=df.reset_index(drop=True)
```

```python
indices=pd.Series(df.index,index=df['title'])
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(stop_words='english',max_features=50000,ngram_range=(1,3))
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  if title not in indices:
    return ["Movie Not Found"]
  idx=indices[title]
  sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  #Sorts movie indices based on similarity score (low → high)
  #Reverses order (high → low similarity)
  # Skips the first movie (because it’s the same movie)
  # Takes the next n most similar movies
  similar_idx=sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Inception')
```

```python
import pickle
pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(df,open('df.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
