```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text=text.lower()
  #removing punctuations
  text=re.sub(r'[^\a-zA-Z\s]','',text)
  #removing stopwords
  words=text.split()
  words=[word for word in words if word not in stop_words]
  #lemmatization(reducing words to their base form
  words=[lemmatizer.lemmatize(word) for word in words]
  return ' '.join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
df=df.reset_index(drop=True)
```

```python
indices=pd.Series(df.index,index=df['title'])
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(stop_words='english',max_features=50000,ngram_range=(1,3))
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  if title not in indices:
    return ["Movie Not Found"]
  idx=indices[title]
  sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  #Sorts movie indices based on similarity score (low → high)
  #Reverses order (high → low similarity)
  # Skips the first movie (because it’s the same movie)
  # Takes the next n most similar movies
  similar_idx=sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Inception')
```

```python
from transformers import pipeline

emotion_classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=False
)

def get_mood(text):
    if not isinstance(text, str):
        return None
    try:
        result = emotion_classifier(text[:512])[0]
        return result['label']
    except:
        return None
```

```python
texts = df['overview'].fillna("").apply(lambda x: x[:512]).tolist()
batch_result = emotion_classifier(texts)

df['mood'] = [res['label'] for res in batch_result]
```

```python
def hybrid_recommend(title, top_n=10, topic_boost=0.20, mood_boost=0.10):
    if title not in indices:
        return ["Movie Not Found"]

    idx = indices[title]
    query_topic = df.iloc[idx]['topic_id']
    query_mood = df.iloc[idx]['mood']

    # TFIDF similarities
    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    results = []
    for i, base_score in enumerate(sim_scores):
        if i == idx:
            continue

        movie_topic = df.iloc[i]['topic_id']
        movie_mood = df.iloc[i]['mood']

        score = base_score

        # topic match boost
        if movie_topic == query_topic:
            score += topic_boost

        # mood match boost
        if movie_mood == query_mood:
            score += mood_boost

        results.append((df.iloc[i]['title'], score))

    # sort by final boosted score
    results = sorted(results, key=lambda x: x[1], reverse=True)

    return results[:top_n]
```

```python
hybrid_recommend("Interstellar", top_n=10)
```
